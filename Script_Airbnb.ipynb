```python
# Muestra la cantidad de valores nulos de la variable 'review_scores_rating' en la base 'df_clean'
print(df_clean['review_scores_rating'].isnull().sum())
```

```python
# Definir los intervalos y etiquetas
bins = [20, 26, 32, 38, 44, 50, 56, 62, 68, 74, 80, 86, 92, 98, 104] # Intervalos definidos
labels = ["20-26", "26-32", "32-38", "38-44", "44-50", "50-56", "56-62", "62-68", 
          "68-74", "74-80", "80-86", "86-92", "92-98", "98-104"] # Etiquetas

# Crear los intervalos en el DataFrame
df_clean['intervalos'] = pd.cut(df_clean['review_scores_rating'], bins=bins, labels=labels, right=False)

# Calcular la frecuencia absoluta (f), asegurando que todos los intervalos aparezcan
f = df_clean['intervalos'].value_counts().reindex(labels, fill_value=0)

# Calcular la frecuencia relativa (rf) y en porcentaje (rf%)
rf = f / f.sum()
rf_percent = rf * 100

# Calcular la frecuencia acumulada (cf) y en porcentaje (cf%)
cf = f.cumsum()
cf_percent = rf_percent.cumsum()

# Crear la tabla de frecuencias
tabla_frecuencia = pd.DataFrame({
    'Intervalos': labels,
    'f': f.values,
    'rf': rf.round(2).values,
    'rf(%)': rf_percent.round(2).values,
    'cf': cf.values,
    'cf(%)': cf_percent.round(2).values
})

# Mostrar tabla
print(tabla_frecuencia)
```

## Prueba Chi cuadrado

```python
# Establecer los valores del test chi cuadrado
chi, pval, dof, exp = chi2_contingency(tabla_contingencia)

# chi: valor del test
# pval: p-valor del test
# dof: grados de libertad

# Imprimir los resultados del test
print('p-valor es: ', pval)

# Establecer la significancia del test
significancia = 0.05
p = 1 - significancia

# Establecer el valor crítico
valor_critico = chi2.ppf(p, dof)
print('chi=%.6f, valor crítico=%.6f\n' % (chi, valor_critico))

# Según el valor crítico se toma la decisión de rechazar o no la hipótesis 
if chi > valor_critico:
    print("""A un %.2f de nivel de significancia, se rechaza la hipótesis nula y se acepta H1. 
No son independientes, por lo tanto están correlacionados""" % (significancia))
else:
    print("""A un %.2f de nivel de significancia, no se rechaza la hipótesis nula. 
Son independientes, no tienen relación""" % (significancia))
```

## Anova

```python
# También para ANOVA necesito la variable "tipo_habitación"
# Importar las librerías necesarias según el análisis que se vaya a realizar (calcular p valor de la F)
import scipy.stats as stats
```

```python
# Variables de interés
data_muestra=df_vars2[['price','room_type']]
tamano_grupos = data_muestra.groupby('room_type').count().reset_index()
tamano_grupos
```

```python
# Varianza de los grupos
varianza_grupos = data_muestra.groupby('room_type').std()
varianza_grupos
```

```python
# Crear la estructura de tabla ANOVA 
data = [['SSA', '', '', '', '', ''], ['SSE', '', '', '', '', ''], ['Total', '', '', '', '', '']] 
anova_tabla = pd.DataFrame(data, columns = ['Fuente de variación', 'SS', 'gdl', 'MS', 'F', 'P-valor']) 
anova_tabla.set_index('Fuente de variación', inplace = True)
anova_tabla
```

```python
# Calcular la SSA y actualizamos la tabla
promedio_precio = data_muestra['price'].mean()
SSA = data_muestra.groupby('room_type').count() * (data_muestra.groupby('room_type').mean() - promedio_precio)**2
anova_tabla.at['SSA', 'SS'] = SSA['price'].sum()
anova_tabla.at['SSA', 'SS']
```

```python
# Calculas la SSE y actualizar en la tabla
SSE = (data_muestra.groupby('room_type').count() - 1) * data_muestra.groupby('room_type').std()**2
anova_tabla.at['SSE', 'SS'] = SSE['price'].sum()
anova_tabla.at['SSE', 'SS']
```

```python
# Calcular SST y actualizar en la tabla
SST = SSA['price'].sum() + SSE['price'].sum()
anova_tabla.at['Total', 'SS']= SST
anova_tabla.at['Total', 'SS']
```

```python
# Calcular los grados de libertas
anova_tabla.loc['SSA', 'gdl'] = data_muestra['room_type'].nunique() - 1
anova_tabla.loc['SSE', 'gdl'] = data_muestra.shape[0] - data_muestra['room_type'].nunique()
anova_tabla.loc['Total', 'gdl'] = data_muestra.shape[0] - 1
```

```python
anova_tabla['gdl']['SSA']
```

```python
anova_tabla['gdl']['SSE']
```

```python
anova_tabla['gdl']['Total']
```

```python
# Calcular MS
anova_tabla['MS'] = anova_tabla['SS'] / anova_tabla['gdl']
anova_tabla['MS']
```

Ya con los cálculos de los grados de libertad y con el MS calculado se puede estimar el valor F y el Valor p para la prueba estadística

```python
# Calcular F 
F = anova_tabla['MS']['SSA'] / anova_tabla['MS']['SSE']
anova_tabla.at['SSA','F'] = F
F
```

```python
# Calcular p-valor
anova_tabla.at['SSA','P-valor'] = 1 - stats.f.cdf(F, anova_tabla.at['SSA','gdl'], anova_tabla.at['SSE','gdl'])
anova_tabla.at['SSA','P-valor']
```

```python
# Evaluar la hipótesis con un nivel de confianza del 95% y significancia del 5%
alpha=0.05 #Establecer la significancia
# Evaluar la hipótesis
print("Usar el P-valor para decidir sobre la hipótesis")
if anova_tabla.at['SSA','P-valor'] <= alpha:
    conclusion = "Se rechaza hipótesis nula "
else:
    conclusion = "No se rechaza hipótesis nula "
print(conclusion)
print("Valor F es:", anova_tabla['F']['SSA'], " y el P-valor es:", anova_tabla['P-valor']['SSA'])    
```

El P-valor es menor que el α por lo tanto se puede rechazar la hipótesis nula, esto muestra que las medias entre tipos de habitación son diferentes.
